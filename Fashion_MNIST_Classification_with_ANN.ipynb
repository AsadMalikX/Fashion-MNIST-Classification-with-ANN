```python
# Import necessary libraries
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Load the Fashion-MNIST Dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()
```

```python
# Step 2: Normalize the Data (Scaling pixel values from 0-255 to 0-1)
x_train, x_test = x_train / 255.0, x_test / 255.0  # Normalization improves convergence
```

```python
# Step 3: Define the Model Architecture (ANN)
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),  # Flatten 28x28 image to 1D array (784)
    keras.layers.Dense(128, activation='relu'),  # Fully connected layer with 128 neurons
    keras.layers.Dropout(0.2),  # Dropout to prevent overfitting
    keras.layers.Dense(10, activation='softmax')  # Output layer with 10 neurons (one per class)
])
```

```python
# Step 4: Compile the Model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',  # Suitable for multi-class classification
              metrics=['accuracy'])
```

```python
# Step 5: Train the Model
history = model.fit(x_train, y_train, epochs=10)
```

```python
# Step 6: Evaluate the Model on Test Data
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"\nTest accuracy: {test_acc * 100:.2f}%")
```

```python
# Step 7: Make Predictions
predictions = model.predict(x_test)
```

```python
# Step 8: Visualize Some Predictions
class_names = ["T-shirt", "Trouser", "Pullover", "Dress", "Coat", "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]

def plot_image(index):
    plt.imshow(x_test[index], cmap=plt.cm.binary)
    plt.title(f"True: {class_names[y_test[index]]} | Pred: {class_names[np.argmax(predictions[index])]}")
    plt.axis("off")
    plt.show()
```

```python
for  i in range (10):
  plot_image(i)
```
